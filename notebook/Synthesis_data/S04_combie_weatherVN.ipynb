```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normalize_to_month_start(df, date_col):
    # Chuyển về datetime nếu chưa phải
    df[date_col] = pd.to_datetime(df[date_col])
    # Quy chuẩn: Tháng 12/2024 (bất kể ngày 01 hay 31) -> 2024-12-01
    df[date_col] = df[date_col].dt.to_period('M').dt.to_timestamp()
    return df
```

```python
humidity_data = pd.read_csv(r"data\raw\weather_data\humidity_vietnam_tft.csv")
preciptiation_data = pd.read_csv(r"data\raw\weather_data\precipitation_vietnam_tft.csv")
solar_data = pd.read_csv(r"data\raw\weather_data\solar_radiation_vietnam_tft.csv")
temp_data = pd.read_csv(r"data\raw\weather_data\temp_vietnam_tft.csv")
humidity_data
```

```python
cols_to_mean = ['humidity_HN', 'humidity_DN', 'humidity_HCM']
humidity_data['humidity_mean'] = humidity_data[cols_to_mean].mean(axis=1)
humidity_data.drop(columns=cols_to_mean, inplace=True)
humidity_data = normalize_to_month_start(humidity_data, 'date')
humidity_data
```

```python
cols_to_mean = ['precip_HN', 'precip_DN', 'precip_HCM']
preciptiation_data['precip_mean'] = preciptiation_data[cols_to_mean].mean(axis=1)
preciptiation_data.drop(columns=cols_to_mean, inplace=True)
preciptiation_data = normalize_to_month_start(preciptiation_data, 'date')
preciptiation_data
```

```python
cols_to_mean = ['solar_HN', 'solar_DN', 'solar_HCM']
solar_data['solar_mean'] = solar_data[cols_to_mean].mean(axis=1)
solar_data.drop(columns=cols_to_mean, inplace=True)
solar_data = normalize_to_month_start(solar_data, 'date')
solar_data
```

```python
cols_to_mean = ['HN_temp', 'DN_temp', 'HCM_temp']

temp_data['temp_mean'] = temp_data[cols_to_mean].mean(axis=1)
col_drop = [col for col in temp_data.columns if col not in ['date', 'temp_mean']]
temp_data.drop(columns=col_drop, inplace=True)
temp_data = normalize_to_month_start(temp_data, 'date')
temp_data
```

```python
combie_data = humidity_data.merge(preciptiation_data, on='date', how='inner')
combie_data = combie_data.merge(solar_data, on='date', how='inner')
combie_data = combie_data.merge(temp_data, on='date', how='inner')
#combie_data = combie_data.merge(humidity_data, on='date', how='inner')
combie_data.isnull().sum()
```

```python
combie_data.to_csv(r"data\processed\weather_vietnam_monthly.csv", index=False)
```
