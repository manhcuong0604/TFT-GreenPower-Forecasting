```python
import pandas as pd
def normalize_to_month_start(df, date_col):
    # Chuyển về datetime nếu chưa phải (thêm dayfirst=True để parse đúng ngày/tháng)
    df[date_col] = pd.to_datetime(df[date_col], dayfirst=True)

    # Quy chuẩn: Tháng 12/2024 (bất kể ngày 01 hay 31) -> 2024-12-01
    df[date_col] = df[date_col].dt.to_period('M').dt.to_timestamp()

    # drop hết dữ liệu trước tháng 1/2018
    df = df[df[date_col] >= '2018-01-01']

    return df
```

```python

# df_argentina = normalize_to_month_start(df_argentina, 'date')
electric_data = normalize_to_month_start(electric_data, 'date')
electric_data.head()
```
