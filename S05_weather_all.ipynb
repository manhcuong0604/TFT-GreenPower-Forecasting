```python
import pandas as pd
def normalize_to_month_start(df, date_col):
    # Chuyển về datetime nếu chưa phải
    df[date_col] = pd.to_datetime(df[date_col])
    # Quy chuẩn: Tháng 12/2024 (bất kể ngày 01 hay 31) -> 2024-12-01
    df[date_col] = df[date_col].dt.to_period('M').dt.to_timestamp()
    return df
```

```python
df_argentina = pd.read_csv(r'data\processed\All_weather_data\argentina_weather.csv')
electric_data = pd.read_csv(r'data\raw\ember_data\all_country_data_cleaned.csv')

df_argentina = normalize_to_month_start(df_argentina, 'date')
df_argentina.head()
```

```python
# combine dữ liệu dựa vào cột entity = arena và date
combined_df = pd.concat([electric_data, df_argentina], keys=['electric', 'weather'], axis=1)
combined_df.head()
```
